```python
import sympy as spy
x=spy.Symbol("x")
g=spy.exp(x/2)*spy.sin(x/3)
#primera derivada
f1=spy.diff(g,x)
print(g)
print(f1)
```

```python
#segunda derivada
f2=spy.diff(f1,x)
print(f2)
```

```python
#tercera derivada
f3=spy.diff(f2,x)
print(f3)
```

```python
#cuarta derivada
f4=spy.diff(f3,x)
print(f4)
```

```python
import matplotlib.pyplot as plt
import numpy as np
def f(x):
    return np.exp(x/2)*np.sin(x/3)
def P1(x):
    return x/3
def P2(x):
    return (x/3)+((x**2)/6)
def P3(x):
    return (x/3)+((x**2)/6)+((23*x**3)/618)
```

```python
# Graficar  funciones.
plt.figure(figsize=(10,10))


plt.plot(x, f(x),label="f(x)=e^(x/2)*sen(x/3)")
plt.plot(x, P1(x),label="P1(x)=x/3")
plt.plot(x, P2(x),label="P2(x)=(x/3)+((x^2)/6)")
plt.plot(x, P3(x),label="P3(x)= (x/3)+((x^2)/6)+((23x^3)/618)")


plt.axhline(0, color="black",linestyle=":")
plt.axvline(0, color="black",linestyle=":")
plt.xlim(-6, 11)
plt.ylim(-26, 26)


plt.xlabel("Eje x")
plt.ylabel("Eje y")
plt.title("Polinomio de Maclaurin")
plt.legend(loc = "upper left")
plt.show()
```
