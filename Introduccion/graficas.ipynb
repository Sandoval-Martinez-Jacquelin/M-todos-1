```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Función cuadrática.
def f(x):
    return (x+1)**(1/2)
```

```python
def g(x):
    return 1+(1/(2*x))
```

```python
def h(x):
    return 1+(1/(2*x)) +(-1/(8*x**2))
```

```python
def j(x):
    return 1+(1/(2*x)) +(-1/(8*x**2))+ (1/(16*x**3))
```

```python
# Valores del eje X que toma el gráfico.
x=np.linspace(-5, 5,100)
```

```python
# Graficar ambas funciones.
plt.figure(figsize=(10,10))

plt.plot(x, f(x))
plt.plot(x,g(x))
plt.plot(x,h(x))
plt.axhline(0, color="black")
plt.axvline(0, color="black")
plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.savefig("output.png") # Guardar gráfico como imágen PNG.
# Mostrarlo.
plt.grid(color="black")
plt.xlabel("Eje x")
plt.ylabel("Eje y")
plt.title("Gráfica de $f(x)=x-3^-x$")
plt.show()
#se puede observar que la gráfica de la función f(x) resulta de la suma de las funciones g(x) + h(x) y
```
